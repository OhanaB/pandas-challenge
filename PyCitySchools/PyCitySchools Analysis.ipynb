```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path(R"D:\Ohana - Data Analysis Bootcamp\Classwork\WAUS-VIRT-DATA-PT-03-2023-U-LOLC\02-Homework\04-Data-Analysis-Pandas\PyCitySchools\Resources\schools_complete.csv")
student_data_to_load = Path(R"D:\Ohana - Data Analysis Bootcamp\Classwork\WAUS-VIRT-DATA-PT-03-2023-U-LOLC\02-Homework\04-Data-Analysis-Pandas\PyCitySchools\Resources\students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python

# Local Gorvernment Area (LGA) Summary 

# Calculate the total number of unique schools 
Total_Schools = len(school_data_complete["school_name"].unique())

# Calculates the total number of unique students
Total_Students = len(school_data_complete["Student ID"].unique())

# Total Budget 
Total_Budget = school_data["budget"].sum()

# Average Maths Score 
Avg_Maths_Score = school_data_complete["maths_score"].mean()

# Average Reading Score 
Avg_Reading_Score = school_data_complete["reading_score"].mean()

# % Passing Maths (The percentage of students who passed maths)
Passing_Maths = school_data_complete[school_data_complete["maths_score"] >= 50]["maths_score"].count()
Percentage_Passing_Math = Passing_Maths / Total_Students * 100

# % Passing Read (The percentage of students who passed reading)
Passing_Read = school_data_complete[school_data_complete["reading_score"] >= 50]["reading_score"].count()
Percentage_Passing_Read = Passing_Read / Total_Students * 100

# % Overall Passing (the percentage of students who passed maths and reading)
Overall_Passing = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)]["Student ID"].count()
Percentage_Passing_Overall = Overall_Passing / Total_Students * 100 

# New Data Frame Local Government Area (LGA) Summary
Government_LGA_Summary = pd.DataFrame({
 "Total Schools": [Total_Schools],
 "Total Students": [Total_Students],
 "Total Budget": [Total_Budget],
 "Average Maths Score": [Avg_Maths_Score],
 "Average Reading Score": [Avg_Reading_Score],
 "% Passing Maths": [Percentage_Passing_Math],
 "% Passing Reading Score": [Percentage_Passing_Read],
 "% Overall Passing": [Percentage_Passing_Overall]
})
# Format the Collumns 
Government_LGA_Summary["Total Budget"] = Government_LGA_Summary["Total Budget"].map("${:,.2f}".format)
Government_LGA_Summary["Total Students"] = Government_LGA_Summary["Total Students"].map("{:,}".format)
Government_LGA_Summary
```

```python
# School Summary

# Group the data by school name
by_school = school_data_complete.groupby('school_name')

# Calculate Total Students and Total School Budget for each school
per_school_summary = pd.DataFrame({
    'Total Students': by_school['Student ID'].count(),
    'Total School Budget': by_school['budget'].mean()
})

# Calculate per student budget for each school
per_school_summary['Per Student Budget'] = per_school_summary['Total School Budget'] / per_school_summary['Total Students']

# Formart Collumns adding the dollar sign 
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map('${:,.2f}'.format)
per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map('${:,.2f}'.format)

# Average maths score - Calculates the Average Scores
per_school_summary['Average Maths Score'] = by_school['maths_score'].mean()

# Average reading score - Calculates the Average Scores
per_school_summary['Average Reading Score'] = by_school['reading_score'].mean()

# Calculate % passing maths for each school (the percentage of students who passed maths)
per_school_summary['% Passing Maths'] = by_school.apply(lambda x: (x['maths_score'] >= 50).sum() / x['maths_score'].count() * 100)

# Calculate % passing reading for each school (the percentage of students who passed reading)
per_school_summary['% Passing Reading'] = by_school.apply(lambda x: (x['reading_score'] >= 50).sum() / x['reading_score'].count() * 100)

# Calculate % overall passing for each school (the percentage of students who passed maths and reading)
per_school_summary['% Overall Passing'] = by_school.apply(lambda x: ((x['maths_score'] >= 50) & (x['reading_score'] >= 50)).sum() / x['maths_score'].count() * 100)

# Add school type to the DataFrame
school_type = school_data[['school_name', 'type']]
per_school_summary = pd.merge(per_school_summary, school_type, on='school_name')

# Reorder the columns
columns = ['school_name', 'type', 'Total Students', 'Total School Budget', 'Per Student Budget', 'Average Maths Score', 'Average Reading Score', '% Passing Maths', '% Passing Reading', '% Overall Passing']
per_school_summary = per_school_summary[columns]


# Setting the index to 'school_name'
per_school_summary.set_index('school_name', inplace=True)
per_school_summary = per_school_summary.rename(columns={'type': 'School Type'})


# Diplays School Summary DataFrame
per_school_summary.index.name=None
per_school_summary  
```

```python
# Highest Performing Schools (By % Overalll Passing)
# Sorting the schools by % Overall Passing in descending order and displaying the top 5 rows
top_schools = per_school_summary.sort_values('% Overall Passing', ascending=False).head(5)
top_schools.index.name=None
top_schools
```

```python
# Lowest Performing Schools (By % Overall Passing)
#Sorting the schools by % Overall Passing in ascending order and displaying the top 5 rows.
bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True).head(5)
bottom_schools.index.name=None
bottom_schools
```

```python
# Maths Scores by Year
year_9 = school_data_complete[(school_data_complete["year"] == 9)]
year_10 = school_data_complete[(school_data_complete["year"] == 10)]
year_11 = school_data_complete[(school_data_complete["year"] == 11)]
year_12 = school_data_complete[(school_data_complete["year"] == 12)]

# DataFrame that list the average maths scores for studentd of each year level 
year9_scores = year_9.groupby("school_name").mean(numeric_only=True)["maths_score"]
year10_scores = year_10.groupby("school_name").mean(numeric_only=True)["maths_score"]
year11_scores = year_11.groupby("school_name").mean(numeric_only=True)["maths_score"]
year12_scores = year_12.groupby("school_name").mean(numeric_only=True)["maths_score"]
maths_scores_by_year = pd.DataFrame({"Year 9": year9_scores, "Year 10": year10_scores,
                                "Year 11": year11_scores, "Year 12": year12_scores})

maths_scores_by_year.index.name=None

# Displays DateFrame 
maths_scores_by_year
```

```python
# Readign Scores By Year
year_9 = school_data_complete[(school_data_complete["year"] == 9)]
year_10 = school_data_complete[(school_data_complete["year"] == 10)]
year_11 = school_data_complete[(school_data_complete["year"] == 11)]
year_12 = school_data_complete[(school_data_complete["year"] == 12)]

# DataFrame that lists the average reading score for students of each year level
year9_scores = year_9.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]
year10_scores = year_10.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]
year11_scores = year_11.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]
year12_scores = year_12.groupby(["school_name"]).mean(numeric_only=True)["reading_score"]
reading_scores_by_year = pd.DataFrame({"Year 9": year9_scores, "Year 10": year10_scores,
                                "Year 11": year11_scores, "Year 12": year12_scores})

reading_scores_by_year.index.name=None

#Displays DateFrame 
reading_scores_by_year
```

```python
# Scores by School Spending 
# Identifying the spending bins and labels
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Categorizing spending based on the bins
school_data_complete["Spending Ranges (Per students)"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spending_bins, labels=spending_labels)
spending_maths_scores = school_data_complete.groupby(["Spending Ranges (Per students)"])["maths_score"].mean()
spending_reading_scores = school_data_complete.groupby(["Spending Ranges (Per students)"])["reading_score"].mean()

passing_math_by_spending = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["Spending Ranges (Per students)"])["Student ID"].count()
passing_reading_by_spending = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["Spending Ranges (Per students)"])["Student ID"].count()
overall_passing_by_spending = school_data_complete[(school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50)].groupby(["Spending Ranges (Per students)"])["Student ID"].count()

spending_passing_math = (passing_math_by_spending / school_data_complete.groupby(["Spending Ranges (Per students)"])["Student ID"].count()) * 100
spending_passing_reading = (passing_reading_by_spending / school_data_complete.groupby(["Spending Ranges (Per students)"])["Student ID"].count()) * 100
overall_passing_spending = (overall_passing_by_spending / school_data_complete.groupby(["Spending Ranges (Per students)"])["Student ID"].count()) * 100

# Spending summary DataFrame
spending_summary = pd.DataFrame({"Average Maths Score": spending_maths_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Maths": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing_spending})

# Formatting the columns of the DataFrame
spending_summary["Average Maths Score"] = spending_summary["Average Maths Score"].map("{:.2f}".format)
spending_summary["Average Reading Score"] = spending_summary["Average Reading Score"].map("{:.2f}".format)
spending_summary["% Passing Maths"] = spending_summary["% Passing Maths"].map("{:.2f}".format)
spending_summary["% Passing Reading"] = spending_summary["% Passing Reading"].map("{:.2f}".format)
spending_summary["% Overall Passing"] = spending_summary["% Overall Passing"].map("{:.2f}".format)

# Display the Dataframe
spending_summary
```

```python
# Bin the schools by size
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=size_labels)

# Calculate the average scores and passing rates by school size
size_math_scores = per_school_summary.groupby(["School Size"])["Average Maths Score"].mean()
size_reading_scores = per_school_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby(["School Size"])["% Passing Maths"].mean()
size_passing_reading = per_school_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby(["School Size"])["% Overall Passing"].mean()

# Create the size summary data frame
size_summary = pd.DataFrame({"Average Maths Score": size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Maths": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing": size_overall_passing})

# Format the columns of the DataFrame
size_summary["Average Maths Score"] = size_summary["Average Maths Score"].map("{:.6f}".format)
size_summary["Average Reading Score"] = size_summary["Average Reading Score"].map("{:.6f}".format)
size_summary["% Passing Maths"] = size_summary["% Passing Maths"].map("{:.6f}".format)
size_summary["% Passing Reading"] = size_summary["% Passing Reading"].map("{:.6f}".format)
size_summary["% Overall Passing"] = size_summary["% Overall Passing"].map("{:.6f}".format)

# Display the size summary data frame
size_summary
```
